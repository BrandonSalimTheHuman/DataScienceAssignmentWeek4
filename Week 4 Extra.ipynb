```python
import pandas as pd
import json
import numpy as np

#url of .csv file
url = "https://raw.githubusercontent.com/nnqomariyah/Fundamentals_of_Data_Science/main/week_3/data/movies_metadata.csv"

#read .csv to df
df = pd.read_csv(url)

#there are so many columns I need this to see all of them
pd.set_option('display.max_columns', None)

df.head()
```

```python
#renaming column names to use capitalization and spaces
df.columns = ["Adult", "Belongs to Collection", "Budget", "Genres", "Homepage", "ID", "IMDb ID", "Original Language", "Original Title", "Overview", "Popularity", "Poster Path", "Production Companies", "Production Countries", "Release Date", "Revenue", "Runtime (minutes)", "Spoken Languages", "Status", "Tagline", "Title", "Video", "Vote Average", "Vote Count"]

#making the index start from 1 because it makes more sense that way
df.index = range(1, 45467)

#giving name to index
df.index.name = "No."

#resizing the max column width so I can see the full entries
pd.set_option('display.max_colwidth', None)

#function to get collection
def get_collection(input):
    #NaN counts as floats, which will be left alone. This makes sure it's a string
    if isinstance(input, str):
        #sth I realised is that the collection name is betweent he 5th and 6th apostrophe
        #so I used it to split the string and take the 5th index of the array for the collection name
        #however there were some weird ones which has a collection of like '0.05234' or sth like that, as a string
        #so I make sure it has a length of at least 6 before taking the 5th index
        parts = input.split("'")
        if len(parts) >= 6:
            return parts[5]
        else:
            #I don't think a decimal should be in the collections column at all, so I'll turn it into NaN instead
            return np.nan  

# Apply the function to the column
df['Belongs to Collection'] = df['Belongs to Collection'].apply(get_collection)

df.head()
```

```python
#thousand separator for budget
def thousand_separator(x):
    #seems like the numbers are actually strings
    if isinstance(x, str):
        #converting the strings into integers
        try:
            num_x = int(x)
        #for some reason, there are 3 .jpgs in the budgets column. I'll change those into NaN because it makes no sense to have .jpgs here
        except ValueError:
            return np.nan
        #it makes absolutely no sense that some of these movies have a zero dollar budget
        #keeping it in will disrupt data like median, mean, and maybe even mode
        #so, I'll turn them into missing values
        if num_x == 0:
            return np.nan
        else:
            #return the formatted one with thousand separators
            formatted_num = "$" + format(num_x,",")
        return formatted_num
        
#applying the function to the budgets column
df['Budget'] = df['Budget'].apply(thousand_separator)

df.head()
```

```python
#getting the genres from the "Genres" column
def get_genre(input):
    #makes sure it's a string
    if isinstance(input, str):
        #split it based on '
        parts = input.split("'")
        #the target is on index 5, so make sure the length is at least 6
        if len(parts) >= 6:
            #empty list because there can be multiple genres
            genre_list = []
            #first genre is at index 5
            i = 5
            #while there are more genres
            while i < len(parts):
                #add the genre to the list
                genre_list.append(parts[i])
                #the next genre is always 6 indexes away because of the way the data is formatted
                i += 6
            #variable to hold the result
            genre_results = ""
            for genre in genre_list:
                #add the genre to the string, then add a comma and a space so it looks neat
                genre_results = genre_results + genre + ", "

            #the last genre will still have the comma and the space, so remove the last two characters
            genre_results = genre_results[:-2]

            #return the string
            return genre_results
        else:
            #inputs that don't fit the format of the others are probably errors, so I'll make them NaN
            return np.nan
        

#applying the function to the Genres column
df['Genres'] = df['Genres'].apply(get_genre)

df.head()
```

```python
#capitalizing the language. "en" doesn't look right, I think it should be "EN"
def capitalize(input):
    #some of them are missing, and NaN are floats, so I need to check for strings
    if isinstance(input, str):
        return input.upper()
    
#applying the function to the Original Language column
df['Original Language'] = df['Original Language'].apply(capitalize)

df.head()
```

```python
#getting production company from "Production Companies" column

#function to get production companies
def get_production_companies(input):
    #make sure it's string
    if isinstance(input, str):
        #split using '
        parts = input.split("'")
        #company starts at index 3, so make sure the length is at least 4
        if len(parts) >= 4:
            #create empty list since there can be multiple companies
            companies_list = []
            #starts at index 3
            i = 3
            #while there are still more companies, add it to the list
            while i < len(parts):
                companies_list.append(parts[i])
                #companies are 6 indexes apart
                i += 6
            #variable to hold result
            companies = ""
            for company in companies_list:
                #add company to the string, then a comma and a space
                companies = companies + company + ", "

            #the last company still has the comma and the space, so remove the last two characters
            companies = companies[:-2]

            #return string
            return companies
        else:
            #inputs that don't fit the format of the others are probably errors, so I'll make them NaN
            return np.nan  
            
#applying the function to the Production Companies column
df['Production Companies'] = df['Production Companies'].apply(get_production_companies)

df.head()
```

```python
#getting production countries from "Production Countries" column

#function to get production coutnries
def get_production_countries(input):
    #make sure it's a string
    if isinstance(input, str):
        #split using '
        parts = input.split("'")
        #the first country starts at index 7, so make sure the length is at least 8
        if len(parts) >= 8:
            #create empty list
            countries_list = []
            #country starts at index 7
            i = 7
            #while there are still countries, add it to the list
            while i < len(parts):
                countries_list.append(parts[i])
                #countries are 8 indexes apart
                i += 8
            #variable to hold result
            countries = ""
            #add each country to the list and then add a comma and a space
            for country in countries_list:
                countries = countries + country + ", "

            #the last country still has the comma and the space, so remove the last two characters
            countries = countries[:-2]

            #return result
            return countries
        else:
            #inputs that don't fit the format of the others are probably errors, so I'll make them NaN
            return np.nan  
            
            
    
#applying the function to the Production Countries column
df['Production Countries'] = df['Production Countries'].apply(get_production_countries)

df.head()
```

```python
#thousand separators for revenue
def thousand_separator_for_revenue(x):
    #for some reason, despite the budget previously being string, this one is float
    if isinstance(x, float):
        #converting the floats into integers because I saw the floats end with just .0 and integers with no trailing .0 looks neater
        try:
            num_x = int(x)
        #not sure if there are weird entries, but here, just in case
        except ValueError:
            return np.nan
        #return the formatted one with thousand separators
        formatted_num = "$" + format(num_x,",")
        return formatted_num
            

#applying the function to the Revenue column
df['Revenue'] = df['Revenue'].apply(thousand_separator_for_revenue)

df.head()
```

```python
#getting the spoken languages from the "Spoken Languages" column

#function to get spoken_languages
def get_spoken_languages(input):
    #make sure it's a string
    if isinstance(input, str):
        #split using '
        parts = input.split("'")
        #the first language starts at index 7, so make sure the length is at least 8
        if len(parts) >= 8:
            #create empty list
            languages_list = []
            #language starts at index 7
            i = 7
            #while there are still languages, add it to the list
            while i < len(parts):
                languages_list.append(parts[i])
                #languages are 8 indexes apart
                i += 8
            #variable to hold result
            languages = ""
            #add each language to the result and add a comma and a space
            for language in languages_list:
                languages = languages + language + ", "

            #the last language still has the comma and the space, so remove the last two characters
            languages = languages[:-2]

            #return result
            return languages
        else:
            #inputs that don't fit the format of the others are probably errors, so I'll make them NaN
            return np.nan  

    
#applying the function to the Spoken Languages column
df['Spoken Languages'] = df['Spoken Languages'].apply(get_spoken_languages)

df.head()
```

```python
#as far as I can tell, the runtime all end with .0, so I'll change them into integers to make it look nicer
def turn_to_integer(x):
    #NaNs are also floats, but can't be turned into integers, so I need to make sure x isn't NaN
    if pd.notna(x):
        num_x = int(x)
        return num_x

#applying the function to the Runtime column
df['Runtime (minutes)'] = df['Runtime (minutes)'].apply(turn_to_integer)

#convert the column type to integers or it will still use floats
df['Runtime (minutes)'] = df['Runtime (minutes)'].astype('Int64')

df.head()
```

```python
#changing the max column width so each row isn't too high, now that I don't have to see the full entry for everything
pd.set_option('display.max_colwidth', 100)

df.head()
```

```python
#seems like the votes also end with .0, so I'll change them into integers too

#applying the function to the Runtime column
df['Vote Count'] = df['Vote Count'].apply(turn_to_integer)

#convert the column type to integers or it will still use floats
df['Vote Count'] = df['Vote Count'].astype('Int64')

df.head()
```
