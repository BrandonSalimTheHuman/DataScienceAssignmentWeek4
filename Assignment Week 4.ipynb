```python
#import stuff
import pandas as pd
import numpy as np
```

```python
#reading csv
df = pd.read_csv("uscs_map_incidence_all.csv")
df2 = pd.read_csv("uscs_map_death_all.csv")
```

```python
#splitting "Range" column into "Range from" and "Range to"
#Typo in the word 'to' is intentional
df[['Range from (Incidence)', 'Range too (Incidence)']] = df['Range'].str.split(' to ', expand=True).astype(float)

#just renaming the rate column
df.rename(columns={'Rate': 'Rate (Incidence)'}, inplace=True)

#removing the original Range column
df.drop(columns=['Range'], inplace=True)

#placing the Rate column back in its original position
#extract columns into list
columns = df.columns.tolist()

#temporarily store the second column
tmp_column = columns.pop(1)

#insert it back in correct position
columns.insert(3, tmp_column)

#assign the correct columns back to df
df = df[columns]
```

```python
#splitting "Range" column into "Range from" and "Range to" for second database
df2[['Range from (Death)', 'Range to (Death)']] = df2['Range'].str.split(' to ', expand=True).astype(float)

#just renaming the rate column
df2.rename(columns={'Rate': 'Rate (Death)'}, inplace=True)

#removing the original Range column from second database
df2.drop(columns=['Range'], inplace=True)

#placing the Rate column back in its original position
#extract columns into list
columns = df2.columns.tolist()

#temporarily store the second column
tmp_column = columns.pop(1)

#insert it back in correct position
columns.insert(3, tmp_column)

#assign the correct columns back to df
df2 = df2[columns]
```

```python
#applying random missing values
np.random.seed(50)
col_name=['Range from (Incidence)', 'Range too (Incidence)', 'Rate (Incidence)']

mask = np.random.choice([True, False, False, False], size=df[col_name].shape)
mask[mask.all(1),-1] = 0
df[col_name]=df[col_name].mask(mask)
```

```python
#applying random missing values to second database
np.random.seed(30)
col_name=['Range from (Death)', 'Range to (Death)', 'Rate (Death)']

mask = np.random.choice([True, False, False, False], size=df2[col_name].shape)
mask[mask.all(1),-1] = 0
df2[col_name]=df2[col_name].mask(mask)
```
